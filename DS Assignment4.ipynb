#### 	Final cleaned dataset with accurate billing info. ####

```python
import pandas as pd
```

```python
# Load dataset
df = pd.read_csv("D:/DS.24th July Banner Batch/Billing_Data.csv")
df
```

```python
# Step 1: Keep relevant columns
df = df[['PatientID', 'InsuranceCovered', 'FinalAmount']]
df
```

```python
# Step 2: Remove duplicates
df = df.drop_duplicates(subset='PatientID')
df
```

```python
# Step 3: Fill missing FinalAmount with mean
df['FinalAmount'] = df['FinalAmount'].fillna(df['FinalAmount'].mean())
df
```

```python
# Step 4: Add insurance info (example logic: 50% insurance if covered)
df['InsuranceCovered'] = [True, False, True, True, False]  # Example values
df['InsuranceAmount'] = df['FinalAmount'] * 0.5 * df['InsuranceCovered'].astype(int)
df
```

```python
# Step 5: Calculate final amount
df['FinalAmount'] = df['FinalAmount'] - df['InsuranceAmount']
df
```

```python
# Final cleaned dataset
print(df)
print
```

```python
import pandas as pd
```

```python
# Load datasets
patients = pd.read_csv("D:\DS.24th July Banner Batch\Patient_Data.csv")
billing = pd.read_csv("D:/DS.24th July Banner Batch/Billing_Data.csv")
patients
```

```python
# Merge on PatientID
df = pd.merge(patients, billing, on="PatientID", how="inner")
df
```

```python
# Drop administrative columns
df = df.drop(columns=['ReceptionistID', 'CheckInTime'], errors='ignore')
df
```

```python
# Remove duplicates
df = df.drop_duplicates(subset='PatientID')
df
```

```python
# Fill missing BillAmount with mean
df['BillAmount'] = df['BillAmount'].fillna(df['BillAmount'].mean())
df
```

```python
# Add new billing columns (example logic)
df['InsuranceCovered'] = [True, False, True, True, False]  # example values
df['InsuranceAmount'] = df['BillAmount'] * 0.5 * df['InsuranceCovered'].astype(int)
df['FinalAmount'] = df['BillAmount'] - df['InsuranceAmount']
df
```

#### 	Ability to perform further analytics on department-wise revenue or doctor performance. ####

#### 1 Department-wise Revenue ####

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Assuming df is your final cleaned dataset
dept_revenue = df.groupby('Department')['FinalAmount'].sum().reset_index()

print(dept_revenue)
```

```python
# Optional: visualize
plt.figure(figsize=(8,5))
sns.barplot(x='Department', y='FinalAmount', data=dept_revenue, palette="viridis")
plt.title("Department-wise Revenue")
plt.ylabel("Total Revenue")
plt.show()
```

#### 2 Doctor-wise Performance ####

```python
doctor_revenue = df.groupby('Doctor')['FinalAmount'].sum().reset_index()
doctor_revenue
```

```python
# Optional: visualize
plt.figure(figsize=(8,5))
sns.barplot(x='Doctor', y='FinalAmount', data=doctor_revenue, palette="magma")
plt.title("Doctor-wise Revenue")
plt.ylabel("Total Revenue")
plt.xticks(rotation=30)
plt.show()
```
